```python
from google_play_scraper import reviews
import pandas as pd

result, continuation_token = reviews(
    'notion.id',
    lang='ko',
    country='kr',
    count=10
)

data = []

for r in result:
    data.append({
        "user": r["userName"],
        "review": r["content"],
        "score": r["score"],
        "date": r["at"]
    })

df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv("notion.id_reviews.csv", index=False, encoding="utf-8-sig")
```
